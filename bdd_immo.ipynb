# Supression des colonnes inutiles

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
df_clean = df[df["Type local"]=="Appartement"]
df_clean = df_clean[df_clean["Nature mutation"]=="Vente"]
```

```python
df_clean
```

# Formattage des valeurs 

```python
df_clean["Date mutation"] = pd.to_datetime(df_clean["Date mutation"],format="%d/%m/%Y")
```

```python
df_clean["Valeur fonciere"] = df_clean["Valeur fonciere"].str.replace(',','.')
df_clean["Valeur fonciere"] = df_clean["Valeur fonciere"].astype(float)
```

```python
for column in df_clean.columns:
    if "Surface Carrez" in column:
        df_clean[column] = df_clean[column].str.replace(',', '.').astype(float)

surface_carrez_columns = [column for column in df_clean.columns if "Surface Carrez" in column]
df_clean['Surface Carrez'] = df_clean[surface_carrez_columns].sum(axis=1)

df_clean.drop(surface_carrez_columns, axis=1, inplace=True)
```

```python
df_clean["Commune"] = df_clean["Commune"].astype("string")
```

```python
df_clean["Type local"] = df_clean["Type local"].astype("string")
```

```python
df_clean["Code departement"] = df_clean["Code departement"].astype("string")
```

```python
df_clean.info()
```

```python
df_clean.isna().sum()
```

```python
df_clean = df_clean.dropna(subset = "Surface Carrez")
# df = df.dropna(subset = "No voie")
df_clean = df_clean.loc[df["Nature mutation"] == "Vente"]
df_clean = df_clean.drop(["Code service sages", "Reference document", "1 Articles CGI","2 Articles CGI",
        "3 Articles CGI","4 Articles CGI","5 Articles CGI","Section","No disposition",
         "Nature culture","Nature culture speciale","Identifiant local","Prefixe de section","2eme lot",
           "3eme lot","4eme lot","5eme lot","No Volume","1er lot","No plan","Surface terrain","Surface reelle bati","Nombre de lots","B/T/Q",
             "Code type local","Nature mutation"],axis = 1)
df_clean = df_clean.reset_index(drop = True)
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.metrics import mean_squared_error
```

```python
X = df_clean.drop(columns=["Valeur fonciere","Date mutation","Type de voie","Voie","Commune","Type local","Code voie", "Code commune","Code departement"])
y = df_clean["Valeur fonciere"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X.isna().sum()
```

```python
baseline = LinearRegression()
```

```python
model = baseline.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
train_sizes, train_scores, test_scores = learning_curve(model, X, y, cv=5, scoring='neg_mean_squared_error')
```

```python
train_scores_mean = -np.mean(train_scores, axis=1)
test_scores_mean = -np.mean(test_scores, axis=1)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores_mean, label='Train')
plt.plot(train_sizes, test_scores_mean, label='Test')
plt.xlabel('Taille de l\'échantillon d\'apprentissage')
plt.ylabel('Erreur quadratique moyenne')
plt.title('Courbe d\'apprentissage')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

# Prédiction sur l'ensemble de test
predictions = model.predict(X_test)

# Calcul de l'erreur quadratique moyenne
mse = mean_squared_error(y_test, predictions)

# Affichage du score
print(f"Erreur quadratique moyenne (MSE) : {mse}")
```

```python
from sklearn.metrics import r2_score

# Calcul du coefficient de détermination
r2 = r2_score(y_test, predictions)

# Affichage du score R²
print(f"Coefficient de détermination (R²) : {r2}")
```

```python
RMSE = np.sqrt(mse)
RMSE
```

# Création des 4 tables

```python
vente = df_clean[["Date mutation", "Valeur fonciere", "id_vente"]]
vente
```

```python
df_clean["Code departement"] = df_clean["Code departement"].astype("string")
df_clean["Code commune"] = df_clean["Code commune"].astype("string")
df_clean["cod_dep_commune"] = df_clean["Code departement"] + df_clean["Code commune"]
bien = df_clean[['No voie','Type de voie','Voie','Code postal','Commune',"Code departement","Code commune",'Surface Carrez du 1er lot','Type local','Nombre pieces principales']]
bien = bien.drop_duplicates().reset_index()
bien["id_bien"] = bien.index+1
bien["id_codedep_codecommune"] = df_clean["Code departement"] + "_" + df_clean["Code commune"]
bien
```

# Table vente

```python
vente = df_clean.merge(bien)
vente = vente[["id_vente","id_bien","Date mutation","Valeur fonciere"]]
vente
```

# Table region

```python
departement = pd.read_csv("departments.csv", delimiter = ",")
departement= departement[["region_code", "code","name"]]
departement
```

```python
regions = pd.read_csv("regions.csv", delimiter = ",")
regions = regions[["code","name"]]
regions = regions.rename(columns = {"name" : "Nom_regroup", "code": "region_code"})
regions["id_region"] = regions.index+1
regions
```

```python
departements = departement.merge(regions)
region = departements.drop(["region_code"],axis = 1)
#region = region.rename(columns = {"name":"Nom_region"})
region = region.rename(columns = {"code":"Code departement"})
region
```

```python
region_final = region.drop(["Code departement"],axis = 1)
region_final = region_final.drop_duplicates(subset = "Nom_regroup")
region_final
```

# Table commune

```python
commune = bien[["Code postal", "Commune", "Code commune","Code departement","id_codedep_codecommune"]]
commune = commune.merge(region)
commune = commune.drop_duplicates(subset = "Commune")
commune = commune.drop(["name","Nom_regroup"],axis = 1)
commune
```
